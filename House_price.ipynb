**HOUSE PRICE PREDICTION**

**OBJECTIVE**
- To develop a robust and accurate machine learning model for predicting house prices.
- This model will leverage a variety of features and advanced algorithms to provide reliable estimates of property values.


**IMPORT LIBRARY**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**IMPORT DATA**

```python
df = pd.read_csv('/houseprices.csv')
```

```python
y=df['SalePrice']
x=df[['GrLivArea',  'FullBath', 'BedroomAbvGr']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print('Mean Squared Error:',mse)
print('R2 Score:',r2)
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Price')
plt.show()
```

```python
nd=pd.DataFrame({'GrLivArea': [3000],  'FullBath':[3], 'BedroomAbvGr':[5]})
pp=model.predict(nd)
print('The predicted price of the house is: ',pp[0])
```
